```python
import numpy as np
import matplotlib.pyplot as plt
from torch.utils.data import Dataset, DataLoader
from torchvision import datasets, transforms
import torch.nn as nn
import torch.optim as optim
```

```python
from modules import Net 
from train_test import *
from helper import *
```

```python
# define the transform
image_transforms = {
    "train": transforms.Compose([transforms.Resize(256),
                                 transforms.CenterCrop(224),
                                 transforms.RandomRotation(30),
                                 
                                 transforms.RandomHorizontalFlip(),
                                 transforms.ToTensor(),
                                 ]),
    "test": transforms.Compose([transforms.Resize(256),
                                transforms.CenterCrop(224),
                                transforms.ToTensor()
                                ])
        }
        
```

```python
# define dataset from folder and data loader
batch_size = 64
data = {"train": datasets.ImageFolder("data/train",image_transforms["train"]),
        "test": datasets.ImageFolder("data/test",image_transforms["test"])}

dataloader = {"train": DataLoader(data["train"],batch_size = batch_size ,shuffle=True),
              "test": DataLoader(data["test"], shuffle=True)}

net = Net()
net = net.to("cuda")
loss_func = nn.CrossEntropyLoss()
```

```python
# if you have trained the net before, load the weightd, otherwise skip this step
state_dict = torch.load("model.pth")
net.load_state_dict(state_dict)
```

```python
# choose the optimizer and train the net
optimizer = optim.Adam(net.parameters(), lr=0.001) 
train_net(20,net,dataloader["train"],loss_func, optimizer)
```

```python
# run the net on the test dataset
test_net(net,dataloader["test"],loss_func)
```

```python
# save the weights of the trained net in model.pth file
torch.save(net.state_dict(),'model.pth')
```
